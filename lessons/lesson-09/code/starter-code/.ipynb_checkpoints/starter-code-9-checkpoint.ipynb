Below is some code to walk through confusion matrices. It'll be useful for working through the Titanic problem.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score
```

Below the ROC curve is based on various thresholds: it shows with a false positive rate (x-axis) ~0, it also expects a true positive rate (y-axis) ~0 (the same, ish, for the top right hand of the figure).

The second chart, which does not play with thesholds, shows the one true TPR and FPR point, joined to 0,0 and 1,1.

The first chart will be more effective as you compare models and determine where the decision line should exist for the data. The second simplifies the first in case this idea of thresholds is confusing.

```python
feature_set = df[['gre', 'gpa', 1, 2, 3,]]
actuals = lm.predict(feature_set) 
probas = lm.predict_proba(feature_set)
plt.plot(roc_curve(df[['admit']], probas[:,1])[0], roc_curve(df[['admit']], probas[:,1])[1])#probas[:,1] selects the second column
```

```python
plt.plot(roc_curve(df[['admit']], actuals)[0], roc_curve(df[['admit']], actuals)[1])
```

Finally, you can use the `roc_auc_score` function to calculate the area under these curves (AUC).

```python
roc_auc_score(df['admit'], probas[:,1])
```

```python
print titanic_data.Survived.value_counts()
#print titanic_data.describe()
```

```python
import seaborn as sns
```

```python
cols_to_keep = ["Survived", 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch']
model_data = titanic_data[cols_to_keep].copy()
model_data['Age'] = model_data['Age'].fillna(method='bfill')
model_data.isnull().sum()
model_data.describe()
sns.pairplot(model_data)
```

```python
sns.factorplot(data=model_data, x="Survived", y="Age")
sns.factorplot(kind='box', data=model_data, x="Survived", y="Age")
```

```python
#model_data.SibSp.apply
model_data['isSibling'] = model_data.SibSp.apply(lambda x: x > 0)
model_data['isChild'] = model_data.Parch.apply(lambda x: x > 0)
#model_data = pd.get_dummies(data = model_data, columns=['Sex', 'Pclass'], drop_first=True)
model_data.head()
sns.pairplot(model_data)
```

```python
feature_set = ['Age', 'isSibling', 'isChild', 'Sex_male', 'Pclass_2', 'Pclass_3']
x = model_data[feature_set]
y = model_data.Survived
```

```python
#LogisticRegression()
lm = LogisticRegression()
lm = LogisticRegression(C=1e12)
lm.fit(x,y)
```

```python
print lm.coef_
print np.exp(lm.coef_) #values for each column
```

```python
# import statsmodels.api as sm
# sm.logit(endog=y, exog=sm.add_constant(x)).fit()
```

```python
actuals = lm.predict(x)
probas = lm.predict_proba(x)
plt.plot(roc_curve(model_data[['Survived']], probas[:,1])[0], roc_curve(model_data[['Survived']], probas[:,1])[1])#probas[:,1] selects the second column
```

```python
plt.plot(roc_curve(model_data[['Survived']], actuals)[0], roc_curve(model_data[['Survived']], actuals)[1])
```

```python
print lm.score(x,y) # accuracy of the model
roc_auc_score(model_data['Survived'], probas[:,1])
```
